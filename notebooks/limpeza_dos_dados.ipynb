```python
import matplotlib.pyplot as plt
import pandas as pd

unidade = r'C:\Users\Jhonatan Pires\Downloads\projeto unidades basicas de saude\data\u.csv'
df_unidade = pd.read_csv(unidade, decimal=',', sep=';')

df_unidade_limpo = df_unidade.dropna(subset=['LATITUDE', 'LONGITUDE'])

ubs_por_uf = df_unidade_limpo['UF'].value_counts().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
ubs_por_uf.plot(kind='bar', color='blue')
plt.title('Quantidade de UBS por UF')
plt.xlabel('UF (Estado)')
plt.ylabel('Número de UBS')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
